```python
import pandas as pd
import pickle

with open('test.pkl', "rb") as f:
    data: pd.DataFrame = pickle.load(f)

# Convert to UTCDateTime if they are not already
from obspy import UTCDateTime

if data['starttime'].dtype == 'object':
    data['starttime'] = data['starttime'].apply(lambda x: UTCDateTime(x))
if data['endtime'].dtype == 'object':
    data['endtime'] = data['endtime'].apply(lambda x: UTCDateTime(x))

print(data['starttime'].dtype)
print(data['endtime'].dtype)

data = data.drop('Unnamed: 0', axis = 1)
data

data.to_pickle('test2.pkl')
```
